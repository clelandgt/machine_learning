```python
## 基础工具
import numpy as np
import pandas as pd
import warnings
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.special import jn
from IPython.display import display, clear_output
import time

warnings.filterwarnings('ignore')
%matplotlib inline

## 模型预测的
from sklearn import linear_model
from sklearn import preprocessing
from sklearn.svm import SVR
from sklearn.ensemble import RandomForestRegressor,GradientBoostingRegressor

## 数据降维处理的
from sklearn.decomposition import PCA,FastICA,FactorAnalysis,SparsePCA

import lightgbm as lgb
import xgboost as xgb

## 参数搜索和评价的
from sklearn.model_selection import GridSearchCV,cross_val_score,StratifiedKFold,train_test_split
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

## 加载数据

```python
data = pd.read_csv('features.csv')
features = [col for col in data.columns if col not in ('price', 'flag', 'SaleID')]

# train_X = data[data['flag']==1][features]
# train_y = data[data['flag']==1]['price']

# test_X = data[data['flag']==0][features]
# test_y = data[data['flag']==0]['price']

train_df = data[data['flag']==1]
```

## 模型

```python
import lightgbm as lgb
import xgboost as xgb
from sklearn.model_selection import train_test_split


train_X, test_X, train_y, test_y = train_test_split(train_df[features], train_df['price'], random_state=666)

xgr = xgb.XGBRegressor(n_estimators=200, learning_rate=0.1, gamma=0, subsample=0.8,\
        colsample_bytree=0.9, max_depth=15)

xgr.fit(train_X, train_y)

pred_test_xgb=xgr.predict(test_X)
score_train = mean_absolute_error(test_y, pred_test_xgb)
print(score_train)
```

## 生成预测数据并导出

```python
# 导出预测值
test_X = data[data['flag']==0][features]

result_df =  data[data['flag']==0][['SaleID']]
result_df['price'] = xgr.predict(test_X)
result_df.to_csv('sample_submit.csv', index=False)
```
