## 决策树训练与可视化

```python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
```

```python
iris = load_iris()
X = iris.data
y = iris.target

print("X shape: ", X.shape)
print("y shape: ", y.shape)
```

```python
dt_clf = DecisionTreeClassifier(max_depth=2)
dt_clf.fit(X, y)
```

```python
import graphviz
from sklearn.tree import export_graphviz

# 法一: 绘制展示在jupyter
dot_data = export_graphviz(dt_clf, feature_names=iris.feature_names, class_names=iris.target_names, filled=True, rounded=True)
graph = graphviz.Source(dot_data) 
graph 
```

```python
# 法二: 保存为图片。 dot -Tpng iris.dot -o iris.png  命令行中将dot转化为png
export_graphviz(dt_clf, feature_names=iris.feature_names, class_names=iris.target_names, out_file='iris.dot', filled=True, rounded=True)
```
